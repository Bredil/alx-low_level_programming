**Identifying and handling the missing values**

```python
#checking missing values
df.isnull().sum()
```

```python
#checking missing values
df.isnull().sum() / df.shape[0] * 100
```

```python
#using KNN imputer for numerical columns to impute missing values
from sklearn.impute import KNNImputer
impute = KNNImputer()

#impute missing values for numerical columns
for i in df.select_dtypes(include="number").columns:
  df[i] = impute.fit_transform(df[[i]])
```

```python
#fill missing values for specific column
for i in ['Polio','Income composition of resources']:
  df[i].fillna( df[i].median(), inplace=True)
```

```python
#check duplicate rows in the dataset
df.duplicated().sum()
#remove duplicate rows
df.drop_duplicates(inplace=True)
#verify after removing duplicates
df.duplicated().sum()
```
